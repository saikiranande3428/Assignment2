```python
import numpy as np
vector1 = np.random.randint(1,20,15)
print(vector1)

vector2=np.reshape(vector1,(3,5))
print(vector2)

print(vector2.shape)

max_elements = np.amax(vector2, axis = 1)
max_elements = max_elements[: , None]
new_arr = np.where(vector2 == max_elements, 0, vector2)
print(new_arr)
```

```python
import pandas as pd
data = pd.read_csv("C:\data.csv")
data.head()
data.describe()
data = data.fillna(data.mean())
print(data[['Maxpulse','Calories']].agg(['min','max','mean','count']))
print(data[(data.Calories < 1000) & (data.Calories > 500)])
print(data[(data.Calories>500) & (data.Pulse < 100)])
df_modified = pd.DataFrame(data, columns = ['Duration', 'Pulse', 'Calories'])
print(df_modified.head())
del data['Maxpulse']
print(data.head())
data['Calories'] = data['Calories'].astype("int")
print(data['Calories'].dtypes)
data.plot.scatter(x = 'Duration', y= 'Calories')
```
